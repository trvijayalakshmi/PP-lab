### Creating a User-defined Module

```python
import mymodule

print("Sum:", mymodule.add(10, 20))
print("Diff:", mymodule.sub(20,40))
print("Product:", mymodule.mul(20,40))
print("Division:", mymodule.div(40,20))
```

```python
import mymodule as mym

print("Sum:", mym.add(10, 20))
print("Diff:", mym.sub(20,40))
print("Product:", mym.mul(20,40))
print("Division:", mym.div(40,20))
```

```python
from mymodule import add, mul

print("Sum:", add(10, 20))
print("Diff:", sub(20,40))
print("Product:", mul(20,40))
print("Division:", div(40,20))
```

```python
from mymodule import *

print("Sum:", add(10, 20))
print("Diff:", sub(20,40))
print("Product:", mul(20,40))
print("Division:", div(40,20))
```
