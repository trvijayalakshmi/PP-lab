```python
def square(a):
    area_s = a*a
    peri_s = 4*a
    return area_s, peri_s
def rectangle(a,b):
    area_p = a*b
    peri_p = 2*(a+b)
    return area_p, peri_p
def triangle(a,b,c):
    area_t = 0.5 * a * b
    peri_t = a + b + c
    return area_t, peri_t
```

```python
import geometry1 as m2

area_t, peri_t = m2.triangle(4,6,5)
print(f"Area and perimeter of a triangle are: {area_t}, {peri_t}")

area_r, peri_r = m2.rectangle(6,4)
print(f"Area and perimeter of a rectangle are: {area_r}, {peri_r}")

area_s, peri_s = m2.square(10)
print(f"Area and perimeter of a square are: {area_s}, {peri_s}")
```
