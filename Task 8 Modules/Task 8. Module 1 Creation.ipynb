### 3. Creating a User-defined Module

```python
def add(a, b):
    return a + b

def sub(a,b):
    return a - b

def mul(a,b):
    return a * b

def div(a,b):
    return a / b
```

```python
import arithmetic

print("Sum:", arithmetic.add(10, 20))
print("Diff:", arithmetic.sub(20,40))
print("Product:", arithmetic.mul(20,40))
print("Division:", arithmetic.div(40,20))
```

```python
import arithmetic as m1

print("Sum:", m1.add(10, 20))
print("Diff:", m1.sub(20,40))
print("Product:", m1.mul(20,40))
print("Division:", m1.div(40,20))
```

```python
from arithmetic import add, mul

print("Sum:", add(10, 20))
print("Diff:", sub(20,40))
print("Product:", mul(20,40))
print("Division:", div(40,20))
```
