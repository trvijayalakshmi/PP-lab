```python
file = open('sample1.txt','r')
num_words = 0
num_vowels = 0
num_spaces = 0
num_lower = 0
num_upper = 0

for line in file:
    num_words += len(line.split())
    for char in line:
        if char.lower() in 'aeiou':
            num_vowels = num_vowels + 1
        if char == ' ':
            num_spaces = num_spaces + 1
        if char.islower():
            num_lower = num_lower + 1
        if char.isupper():
            num_upper = num_upper + 1
        
print(f"Number of words: {num_words}")
print(f"Number of vowels: {num_vowels}")   
print(f"Number of spaces: {num_spaces}")
print(f"Number of lowercase letters: {num_lower}")
print(f"Number of uppercase letters: {num_upper}")
```

```python
file = open('sample1.txt','r') 
word_counts = {}
most_common_word = None
max_occurrences = 0

for line in file:
    words = line.strip().split()
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

        if word_counts[word] > max_occurrences:
            max_occurrences = word_counts[word]
            most_common_word = word
        
print(f"The most common word is '{most_common_word}' with {max_occurrences} occurrences.")
```

### Regular Expressions

```python
import re
# Regular expressions for validation
phone_pattern = r'^[6-9]\d{9}$'   # Valid Indian phone number: 10 digits, starts with 6-9
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w{2,}$'  # General email format

# Read inputs
phone = input("Enter phone number: ")
email = input("Enter email ID: ")

# Validate phone
if re.match(phone_pattern, phone):
    print("Valid Phone Number")
else:
    print("Invalid Phone Number")

# Validate email
if re.match(email_pattern, email):
    print("Valid Email ID")
else:
    print("Invalid Email ID")
```

```python
import re

text = "Please call 67 555-123-4567 for assistance or 987-654-3210 for inquiries."

# Define a regular expression pattern for matching phone numbers
pattern = r'\b\d{3}-\d{3}-\d{4}\b'

# Use re.findall() to find all phone numbers in the text
phone_numbers = re.findall(pattern, text)

# Print the matched phone numbers
for phone in phone_numbers:
    print(phone)
```
