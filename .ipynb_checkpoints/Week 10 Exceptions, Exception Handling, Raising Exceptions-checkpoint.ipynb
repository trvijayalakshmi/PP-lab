## 2. Exception Handling

```python
To detect exceptions, write a code that might cause an error 
and enclose it in a try-except block. The try block contains the 
code that could potentially raise an exception.

When an exception is raised inside the try block, the program 
jumps to the corresponding except block. The except block contains
the code to handle the exception gracefully, allowing the program 
to continue running.
```

```python
a = int(input("enter the first no: "))
b = int(input("enter the second no: "))
try:
    x = a / b  # This will raise a ZeroDivisionError
except:
    print("can't divide by zero")
else:
    print(f"The result is: {x}")
finally:
    print("Execution finished.")
```

```python
s = "123"  ## "123"  "abc"
try:
    result = int(s)
except:
    print(f"Error: '{s}' is not a valid integer!")
else:
    print(f"The integer value is: {result}")
finally:
    print("Execution finished.")
```

```python
my_list = [1, 2, 3]
try:
    value = my_list[7]
except:
    print("Index out of range.")
else:
    print(f"Value: {value}")
finally:
    print("Execution finished.")
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['b']
except:
    print("Key not found in dictionary.")
else:
    print(f"Value: {value}")
finally:
    print("Execution finished.")
```

```python
try:
    x = int("abc")   # This will raise ValueError
    y = 10 / 0       # This would raise ZeroDivisionError
except:
    print("An error occurred!")
```

```python
try:
    x = int("abc")   # This will raise ValueError
    y = 10 / 5       # This would raise ZeroDivisionError
except ZeroDivisionError:
    print("An error occurred!")
```

```python
try:
    x = int("abc")   # This will raise ValueError
    y = 10 / 5       # This would raise ZeroDivisionError
except ValueError:
    print("An error occurred!")
```

```python
try:
    x = int(3)   # This will raise ValueError
    y = 10 / 0       # This would raise ZeroDivisionError
except ValueError:
    print("An error occurred!")
```

### 3. Raising Exceptions Manually

```python
x = -1
if x < 0:
  raise Exception("Sorry, no numbers below zero are allowed")
```

```python
x = "hello"
if not type(x) is int:
  raise ValueError("Only integers are allowed")
```

```python
x = "hello"
if not type(x) is int:
  raise TypeError("Only integers are allowed")
```

```python
try:
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError("Age cannot be negative")
    print("Your age is:", age)
except ValueError as e:
    print("Error:", e)
```

```python
try:  
    a = int(input("Enter a : "))  
    b = int(input("Enter b : "))  
    if b is 0:  
        raise ArithmeticError;  
    else:  
        print("a/b = ",a/b) 
except ValueError:
	print("The values must be numbers")
except ArithmeticError:  
    print("The value of b can't be 0")
```

```python
try:
    raise ValueError("Invalid number entered")
except ValueError as e:
    print("Exception caught:", e)
```

```python
try:
    raise TypeError("Wrong type used", 404, "Extra Info")
except TypeError as e:
    print("Exception caught with arguments:", e.args)  # tuple of arguments
    print("First argument:", e.args[0])
    print("Second argument:", e.args[1])
    print("Third argument:", e.args[2])
```
