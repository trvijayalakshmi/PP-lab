### Inheritance

```python
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
            
poly = Polygon(4)    
poly.inputSides()
poly.dispSides()
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the perimeter
        s = (a + b + c) 
        print('The perimeter of the triangle is %0.2f' %s)
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
        
# Creating an instance of the Triangle class
t = Triangle()

# Prompting the user to enter the sides of the triangle
t.inputSides()

# Displaying the sides of the triangle
t.dispSides()

# Calculating and printing the area of the triangle
t.findArea()
```

```python
class Person:
    def __init__(self, name): self.name = name

class Employee(Person):  # Employee (inherits from Person)
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id  # additional attribute: emp_id

class Manager(Employee):   # Manager (inherits from Employee)
    def __init__(self, name, emp_id, dept):
        super().__init__(name, emp_id)
        self.dept = dept  # additional attribute: dept.

    def display(self):
        print(f"Name: {self.name}, ID: {self.emp_id}, Dept: {self.dept}")

m = Manager("Rajesh", 101, "ECE")  # Create an object
m.display()
```

### Polymorphism

```python
class Shape:
    def area(self):
        print("Area not defined")

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
#         super().area()
        Shape.area(self)
        print(3.14 * self.r * self.r)        
        
class Square(Shape):
    def __init__(self, s):
        self.s = s
    def area(self):
#         super().area()
        Shape.area(self)
        print(self.s * self.s)        
        
c = Circle(5)   # Object creation
c.area()

d = Square(4)   # Object creation
d.area()
```

```python
class Calculator:
    def add(self, a=0, b=0, c=0):
        return a + b + c

c = Calculator()
print(c.add(2, 3))       
print(c.add(2, 3, 4))    
print(c.add(5))          
```

```python
class Display:
    def show(self, value):
        if isinstance(value, int):
            print("Integer:", value)
        elif isinstance(value, str):
            print("String:", value)
        else:
            print("Unknown type:", value)

d = Display()
d.show(10)          # Integer: 10
d.show("Python")    # String: Python
d.show(10.7)
d.show([1, 2, 3]) 
```
