### Programs using if elif else

```python
char = input("Enter a character: ")

if char.isdigit():
    print(f"{char} is a digit.")
elif char.islower():
    print(f"{char} is a lowercase character.")
elif char.isupper():
    print(f"{char} is an uppercase character.")
else:
    print(f"{char} is a special character.")
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if num1 > num2:
    print(num1, "is the largest number.")
else:
    print(num2, "is the largest number.")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("The largest number is:", largest)
```

```python
marks = int(input("Enter your marks (0-100): "))

if marks >= 90:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 50:
    grade = "D"
elif marks >= 40:
    grade = "E"
else:
    grade = "F (Fail)"

print("Your grade is:", grade)
```

```python
year = int(input("Enter a year: "))

if year % 4 == 0:          # First check: divisible by 4
    if year % 100 == 0:    # If divisible by 100, then check further
        if year % 400 == 0:   # If divisible by 400 -> Leap Year
            print(year, "is a Leap Year")
        else:
            print(year, "is NOT a Leap Year")
    else:
        print(year, "is a Leap Year")
else:
    print(year, "is NOT a Leap Year")
```

```python
op = input('enter the operator')
a = int(input('enter the value'))
b = int(input('enter the value'))

if op == "+" or "add":
    result = a + b
elif op == "-" or "sub":
    result = a - b
elif op == "*" or 'mul':
    result = a * b
elif op == "/" or 'div':
    result = a / b if b != 0 else "Error"
else:
    result = "Invalid operation"
    
print(result)
```

```python
p = input('enter the operator')
a, b = 5, 3

match p:
    case '+':
        print(a + b)
    case '-':
        print(a - b)
    case '*':
        print(a * b)
    case '/':
        print(a / b if b != 0 else "Error")
    case _:
        print("Invalid operation")
```

```python
def calculator(op, a, b):
    match op:
        case "add":
            return a + b
        case "sub":
            return a - b
        case "mul":
            return a * b
        case "div":
            # avoid division by zero
            return a / b if b != 0 else "Error! Division by zero"
        case _:
            return "Invalid operation"

# Example usage
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
print("Enter operation: add, sub, mul, div")
operation = input("Your choice: ")

result = calculator(operation, num1, num2)
print("Result:", result)
```
